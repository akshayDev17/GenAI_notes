## `ollama.generate`

### Method signature

```python
ollama.generate?
```

### Simple call

```python
libya_capital_resp = ollama.generate(model='llama3.2', prompt='What\'s the capital of Libya?')
print(libya_capital_resp.response)
```

### Use suffix

```python
japan_national_animal_resp = ollama.generate(model='llama3.2', 
                                     prompt='What\'s the national animal of Japan?',
                                     suffix=' This is just some random suffix.')
print(japan_national_animal_resp.response)
```

### Prompt with large response

```python
essay_500words_streamless_resp = ollama.generate(model='llama3.2', prompt='write me a 500 word essay on Solo Leveling')
print(essay_500words_streamless_resp.response)
```

### Prompt with large, **streamed** response

```python
essay_500words_streamed_resp = ollama.generate(model='llama3.2', prompt='write me a 500 word essay on the USA dropping the gold standard in 1971.', stream=True)
```

```python
type(essay_500words_streamed_resp)
```

So the response is a generator object

```python
num_pieces = 0
for resp in essay_500words_streamed_resp:
    print(resp.response)
    num_pieces += 1

print(f"\n\nTotal pieces = {num_pieces}\n\n")
```

So 604 pieces were generated in the streamed response with each piece being a token(word).
